```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
school_rank = pd.read_csv('Choose_Maryland___Compare_States_-_Education_20241010.csv')
school_rank #reading and displaying data set
```

```python
school_rank = pd.read_csv('Choose_Maryland___Compare_States_-_Education_20241010.csv', usecols = ['State' , 'Graduate Degree Attainment (%)','Annual College Graduates'])
```

```python
school_rank.head(20)
```

```python
#renaming columns to make them easier to refer to/plot
school_rank = school_rank.rename(columns = {'Graduation_Degree_AttainmentAnnual_College_Graduates': 'State','Graduate Degree Attainment (%)': 'Graduate_Degree_Attainment' ,'Annual College Graduates':'Annual_College_Graduates'})
```

```python
school_rank[:5]
```

```python
#Ranking states by annual college graduates descending
#Ranking states by graduate degree attainment descending


school_rank.sort_values(by = ['Annual_College_Graduates', 'Graduate_Degree_Attainment'], ascending = False)[:20]
school_rank.sort_values(by = ['Annual_College_Graduates', 'Graduate_Degree_Attainment'], inplace = True)
school_rank[:20]
```

```python
#Making line graph for graduate degree attainment
plt.figure(figsize = (30,5))
plt.plot(school_rank[:20].State, school_rank[:20].Graduate_Degree_Attainment)
plt.show()
```

```python
#Making line graph for annual college graduates
plt.figure(figsize = (30,5))
plt.plot(school_rank[:20].State, school_rank[:20].Annual_College_Graduates)
plt.show()
```

```python
plt.figure(figsize = (30,5))
plt.scatter(school_rank[:20].State, school_rank[:20].Graduate_Degree_Attainment)
```

```python
plt.figure(figsize = (30,5))
plt.scatter(school_rank[:20].State, school_rank[:20].Annual_College_Graduates)
```
